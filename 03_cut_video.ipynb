```python
import os
import cv2
import math
import shutil
from utils.file_control import get_base_name

# 動画の遅延時間を設定
delay_list = [0, 13.45, 7.5, 17.0, 2.2]
delay_list = [v / 2 for v in delay_list]

# 動画ファイルのパスを取得
video_path_list = [os.path.join("video", f) for f in os.listdir("video") if f.startswith("hirasaki") and f.endswith(".mp4")]
video_path_list.sort()
output_dir = "walk"

# 出力ディレクトリをリセット
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir, exist_ok=True)

# 動画の分割秒数指定
cut_times = [
    [39, 44],
    [64, 69],
    [85, 90],
    [102, 107],
    [119, 123],
    [134, 139],
    [151, 156],
    [166, 172],
    [182, 187],
    [201, 206],
]

# 動画を分割して保存
for j in range(len(video_path_list)):
    print(f"{j}: {video_path_list[j]}")
    for i, (start, end) in enumerate(cut_times):
        video_path = video_path_list[j]
        delay = delay_list[j]
        start -= delay
        end -= delay
        output_path = os.path.join(output_dir, f"{get_base_name(video_path)}_{i}.mp4")

        # OpenCVで動画を読み込み
        cap = cv2.VideoCapture(video_path)
        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        # 開始フレームと終了フレームの計算
        start_frame = max(0, math.ceil(start * fps))
        end_frame = int(end * fps)

        # 動画の書き出し設定
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        # 指定範囲のフレームを取得し書き出し
        current_frame = 0

        while current_frame < end_frame:
            ret, frame = cap.read()
            if not ret:
                break

            # 指定した開始フレームまではスキップ
            if current_frame < start_frame:
                current_frame += 1
                continue

            # 指定範囲内のフレームを書き込み
            out.write(frame)
            current_frame += 1

        # リソースの解放
        cap.release()
        out.release()
        print(f"output: {output_path}")

cv2.destroyAllWindows()
```
