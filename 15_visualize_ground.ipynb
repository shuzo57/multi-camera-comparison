```python
import os

import pandas as pd

import plotly.graph_objects as go
import plotly.express as px

from IPython.display import display
```

```python
csv_dir = "all_combinations"
csv_list = [os.path.join(csv_dir, f) for f in os.listdir(csv_dir) if f.endswith("01234.csv")]
csv_list.sort()
print(csv_list)
```

```python
for csv_path in csv_list:
    df = pd.read_csv(csv_path, index_col=0)
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df.index, y=df["LEFT_ANKLE_z"], mode='lines', name='LEFT_ANKLE'))
    fig.add_trace(go.Scatter(x=df.index, y=df["RIGHT_ANKLE_z"], mode='lines', name='RIGHT_ANKLE'))
    # title = csv_path.split("/")[-1].split(".")[0]
    title = csv_path.split("/")[-1]
    fig.update_layout(title=title)
    display(fig)
```
